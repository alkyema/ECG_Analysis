```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Satwik\Downloads\ECG\Working\cardiologist1_modified.csv")

# Counting rows with all zeros
all_zeros_count = (df.iloc[:, 1:].sum(axis=1) == 0).sum()

# Counting occurrences of each condition
condition_counts = (df != 0).sum()

# Merging all counts into a dictionary
visual_data = condition_counts.to_dict()
visual_data["All Zeros"] = all_zeros_count

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(x=list(visual_data.keys()), y=list(visual_data.values()), palette="magma")

# Labels and title
plt.xlabel("Condition")
plt.ylabel("Count")
plt.title("Distribution of Conditions in the DataFrame")
plt.xticks(rotation=45)
plt.show()
```

```python
zero_patients = df.loc[df.iloc[:, 1:].sum(axis=1) == 0, "Name"].tolist()
print(zero_patients)
```

```python
import random
rand_zero = []
for i in range(50):
    random_value = random.choice(zero_patients)
    print(random_value)
    rand_zero.append(random_value)
print(len(rand_zero))
```

```python
df2 = pd.DataFrame(columns=df.columns)
for i in range(len(rand_zero)):  
    element_add = random.choice(rand_zero)  # Pick a random element
    rand_zero.remove(element_add)  # Remove it from the list
    print("Selected:", element_add)
    
    # Append the selected row to df2
    df2 = pd.concat([df2, df.loc[df["Name"] == element_add]], ignore_index=True)
```
